# Data Report for Categorical Data / Columns
This is a sample report used for our demo to show how you can create a PDF document out of a Jupyter Notebook.
The dataset is open source and containes orders by Indian states and cities with different order status.

```python
### Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

sns.set_style('whitegrid')
sns.set_context("poster")

from typing import List
```

```python
### Import example dataset
df = pd.read_excel("../data/categorical_example_data.xlsx")
```

```python
categorical_columns = ["order_no", "ship_city", "ship_state", "sku", "description", "cod", "order_status"]
```

```python
def custom_bar_plot(x, y, data, hue, order=None, figsize=(16,9), title="Title", ylabel="Y", xlabel="X", xticks=None, xticklabels=None, xrotation=0, color=None, palette='deep'):
    
    plt.figure(figsize = figsize)
    if color:
        sns.barplot(x=x, y=y, hue=hue, data=data, order=order, color=color, alpha=0.8)
    else:
        sns.barplot(x=x, y=y, hue=hue, data=data, order=order, palette=palette, alpha=0.8)
    if xticklabels is None:
        xticklabels = xticks
    plt.xticks(ticks=xticks, labels=xticklabels, rotation=xrotation)
    plt.title(title)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.tight_layout()
    plt.show()
```

The top 5 ship states. Maharashtra is the most prominent state with 0.175 percent of all orders, followed by West Bengal with ~0.11 percent.

```python
## Categorical Barplot - Examples
total_n_rows = df.shape[0]
value_counts = df['ship_state'].value_counts().sort_values(ascending=False).head(5)
custom_bar_plot(x=value_counts.index, y=value_counts.values/total_n_rows, data=None, hue=None)
```

## Cities
Looking at all the cities, Mumbai has the highest share with roughly 10% of all orders.

```python
total_n_rows = df.shape[0]
value_counts = df['ship_city'].value_counts().sort_values(ascending=False).head(6)
custom_bar_plot(x=value_counts.index, y=value_counts.values/total_n_rows, data=None, hue=None, palette="Blues_r")
```
